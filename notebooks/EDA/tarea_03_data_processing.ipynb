```python
import pandas as pd

url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)
df.head()
```

---
## Parte 1 — Exploración y selección

**P1.** Imprime los nombres de todas las columnas y el número total de filas y columnas del DataFrame.

```python
# Tu respuesta aquí
nombre_columna = df.columns
print("El nombre de la primera columna es:\n", nombre_columna)
print("El numero de filas es:\n", df.shape[0])
print("El numero total de columnas es:\n", df.shape[1])
```

**P2.** Usa `.loc` para seleccionar las filas 10 a 15 (inclusive) y solo las columnas `Name`, `Age` y `Survived`.

```python
# Tu respuesta aquí
df.loc[10:15, ["Name", "Age","Survived"]]
```

---
## Parte 2 — Valores nulos y limpieza

**P3.** Cuenta cuántos valores nulos hay en cada columna. ¿Cuáles columnas tienen nulos?

```python
# Tu respuesta aquí
valores_numericos = df.isna().sum()
display("Cantidad de valores numéricos faltantes en cada columna:\n", valores_numericos)
```

**Respuesta:** Las columnas con valores nulos son: "Age", "Cabin" y "Embarked"

**P4.** Rellena los valores nulos de la columna `Age` con la **mediana** de esa columna.  
Verifica que ya no haya nulos en `Age`.

```python
# Tu respuesta aquí
df["Age"] = df["Age"].fillna(df["Age"].mean())
display("Valores de la columna 'Age' después de llenar los valores faltantes:\n", df["Age"].head())
```

---
## Parte 3 — Nuevas columnas y agrupaciones

**P5.** Crea una nueva columna llamada `FamilySize` que sea la suma de `SibSp` + `Parch` + 1 (el pasajero mismo).

```python
# Tu respuesta aquí
FamilySize = df["SibSp"] + df["Parch"] + 1
df["FamilySize"] = FamilySize
display("Las primeras filas de la nueva columna 'FamilySize':\n", df[["SibSp", "Parch", "FamilySize"]].head())
```

**P6.** Usando `groupby`, calcula la **tasa de supervivencia promedio** (columna `Survived`) por clase (`Pclass`).  
¿Qué clase tuvo mayor probabilidad de sobrevivir?

```python
# Tu respuesta aquí
tasa_supervivencia_prom = df.groupby("Pclass")["Survived"].mean()
display("Tasa de supervivencia promedio por clase:\n", tasa_supervivencia_prom)
```

**Respuesta:** La clase con mayor supervivencia fue: la primera clase

**P7.** Calcula la edad promedio y la tarifa (`Fare`) promedio, agrupando por `Sex` y `Pclass`.  
Usa `.groupby()` con `.agg()`.

```python
# Tu respuesta aquí
edad_tarifa_prome = df.groupby("Pclass")[["Age", "Fare"]].mean()
display("Edad y tarifa promedio por clase:\n", edad_tarifa_prome)
```

---
## Parte 4 — Ordenamiento y duplicados

**P8.** Ordena el DataFrame por `Fare` de mayor a menor y muestra las primeras 5 filas.  
¿Cuánto pagó el pasajero más caro?

```python
# Tu respuesta aquí
ordenar_tarifa = df.sort_values(by="Fare", ascending=False)
display("Los pasajeros ordenados por tarifa de mayor a menor:\n", ordenar_tarifa[["Name", "Fare"]].head())
```

**P9.** Verifica si hay filas duplicadas en el DataFrame. ¿Cuántas hay?

```python
# Tu respuesta aquí
filas_duplicadas = df.duplicated().sum()
display("Número de filas duplicadas en el DataFrame:\n", filas_duplicadas)
```
