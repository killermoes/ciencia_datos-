**P6.** Crea los siguientes arreglos e imprime su `shape` y `dtype`:  
- Un arreglo con los valores `[1.5, 2.5, 3.5, 4.5]`  
- Una matriz de ceros de 3 filas × 4 columnas  
- Un arreglo con 6 valores igualmente espaciados entre 0 y 1 (usando `linspace`)

```python
# Tu respuesta aquí
array_v = np.array([1.5, 2.5, 3.5, 4.5])
matriz = np.zeros((3, 4))
array_lins = np.linspace(0, 1 , 6)

print("* Primer arreglo es:\n", array_v)
print(array_v.dtype, array_v.shape)
print("-----------------------------------")
print("* Matriz es:\n", matriz)
print(matriz.dtype, matriz.shape)
print("-----------------------------------")
print("* Arreglo lineal es:\n", array_lins)
print(array_lins.dtype, array_lins.shape)
```

**P7.** Dado el arreglo `a = np.arange(1, 13)`, usa `reshape` para convertirlo en una matriz de 3 × 4.  
Luego imprime el valor máximo, el valor mínimo y la posición (índice) del máximo.

```python
a = np.arange(1, 13)

# Tu respuesta aquí
matriz_corre = a.reshape(3, 4)
print("* Matriz 3 x 4:\n", matriz_corre)
print("-----------------------------------")
print(f"* El numero minimo es: {matriz_corre.min()}\n* El numero maximo es: {matriz_corre.max()}\n* La posición del valor maximo es: {matriz_corre.argmax()}")
```

**P9.** Dado el arreglo `notas = np.array([8.5, 6.0, 9.2, 4.3, 7.8, 5.5, 10.0, 3.1])`,  
usa **indexación booleana** para obtener solo las notas mayores o iguales a 6. Imprime el resultado.

```python
notas = np.array([8.5, 6.0, 9.2, 4.3, 7.8, 5.5, 10.0, 3.1])

notas_m = notas[notas >= 6]
print("Las notas mayores o iguales a 6.0 son:", notas_m)
```
