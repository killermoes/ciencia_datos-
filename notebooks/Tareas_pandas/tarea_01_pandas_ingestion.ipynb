**¿Cuántas personas tienen la palabra Chief en el título de su trabajo?** Apoyese en los métodos para manipular strings y en el método `.apply()`

```python
# Copie el código aca
personas_chief = df[df["JobTitle"].apply(lambda x: "chief" in x.lower())]
cantidad_chief = personas_chief.shape[0]
print(f"La cantidad de personas con 'chief' en su título de trabajo es: {cantidad_chief}")
```

Este código es valido, se recomienda hacer el código mas simple y agregar typing y documentación.

```python
def contar_chief(job_title):
    return "chief" in job_title.lower()
print("La cantidad de personas con 'chief' en su título de trabajo es:", df["JobTitle"].apply(contar_chief).sum())
```

**Bono: ¿Existe una correlación entre la longitud del nombre del Título del trabajo y el Salario?**

crea una nueva columna donde se guarde la longitud del titulo del trabajo. tip: usar el método `.apply()`

```python
# Copie el código aca
df["Long_JobTitle"] = df["JobTitle"].apply(len)
display(df[["JobTitle" , "Long_JobTitle"]].head())
```

Utilice el método `.corr()` de Pandas para calcular la correlacion

```python
# Copie el código aca
correlacion = df[["Long_JobTitle", "TotalPayBenefits"]].corr()
display("La correlación entre la longitud del título de trabajo y el salario total es:\n", correlacion)
```
